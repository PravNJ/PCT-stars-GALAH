```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
plt.rcParams["figure.figsize"] = [16,9]
sns.set_style('whitegrid')
```

```python
TIME = np.arange(0, 2*np.pi,0.001)
sine_amplitude = np.sin(TIME)
sine_amplitude_neg = (-1)*np.sin(TIME)
```

```python
f, ax = plt.subplots(figsize=(16,9))
ax.plot(TIME, sine_amplitude)
ax.plot(TIME,sine_amplitude_neg)
plt.show()
```

```python
import fastdtw
import scipy.spatial.distance as sd
```

```python
def my_fastdtw(spec1, spec2):
    return fastdtw.fastdtw(spec1, spec2)[0]
```

```python
arr = np.array([sine_amplitude, sine_amplitude_neg])
```

```python
distance_matrix = sd.pdist(arr, my_fastdtw)
```

```python
distance_matrix
```
