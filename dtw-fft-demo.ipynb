```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
plt.rcParams["figure.figsize"] = [16,9]
sns.set_style('whitegrid')
```

```python
TIME = np.arange(0, 2*np.pi,0.001)
sine_amplitude = np.sin(TIME)
sine_amplitude_neg = (-1)*np.sin(TIME)
```

```python
f, ax = plt.subplots(figsize=(16,9))
ax.plot(TIME, sine_amplitude)
#ax.plot(TIME,sine_amplitude_neg)
plt.show()
```

```python
from scipy.fft import fft, ifft, fftfreq
```

```python
N = len(TIME)
T= 0.001
```

```python
sine_amplitude_fft = fft(sine_amplitude)
xf = fftfreq(N,T)
```

```python
f, ax = plt.subplots(figsize=(16,9))
ax.plot(xf,sine_amplitude_fft)
#ax.plot(TIME,sine_amplitude_neg)
plt.show()
```
